```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# prameter yg digunakan
n = 1000000
t = np.logspace(np.log10(10),np.log10(500),n)

A = [  3, 6, 1.5, 1.5 ]
d = [ .004, .001, .002, .0015 ]
f = [   3, 1, 4, 2 ]

# Membuat pasangan x dan y
x = A[0]*np.sin(t*f[0])*np.exp(-d[0]*t) + A[1]*np.sin(t*f[1])*np.exp(-d[1]*t)
y = A[2]*np.sin(t*f[2])*np.exp(-d[2]*t) + A[3]*np.sin(t*f[3])*np.exp(-d[3]*t)

# Menampilkan plot nya
plt.plot(x,y,'y',linewidth=.2)
plt.axis('off')
plt.show()
```

```python
# Melihat komponen satu per satu
plt.plot(np.sin(t*2*f[0])*np.exp(-d[0]*0.5*t),linewidth=.5)
plt.show()
```

```python
# menampilakn komponen dari rumus kedua
plt.plot(np.sin(t*f[2])*np.exp(-d[2]*t) + A[3]*np.sin(t*f[3])*np.exp(-d[3]*t),linewidth=.3)
plt.show()
```

```python
# menampilkan komponen dari rumus
plt.plot(np.sin(t*f[1])*np.exp(-d[1]*t),linewidth=.3)
plt.show()
```
